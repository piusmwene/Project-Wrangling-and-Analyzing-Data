### Tidiness Issues

#### Define.

Replace repeated columns with one column known as dog_level

```python
#replace the NaN  of every dog_level with empty space.
levels = ['doggo','pupper', 'floofer', 'puppo' ]
for level in levels:
       df1_archive_enhanced_clean[level] = df1_archive_enhanced_clean[level].replace(np. nan,'',regex=True)
```

```python
# combine dataset using concat and also drop the old columns for four dog level such as doggo, floofer, pupper and puppo. Fill the NaN with empty space
df1_archive_enhanced_clean['dog_level'] = df1_archive_enhanced_clean.doggo.str.cat(df1_archive_enhanced_clean.floofer).str.cat(df1_archive_enhanced_clean.pupper).str.cat(df1_archive_enhanced_clean.puppo)
df1_archive_enhanced_clean.drop(columns =['doggo', 'floofer','pupper', 'puppo'], axis =1, inplace=True)
df1_archive_enhanced_clean['dog_level'] = df1_archive_enhanced_clean['dog_level'].replace('', np.nan)
```

```python
#drop the rows where two dog level combined
df1_archive_enhanced_clean.dog_level.replace('doggopupper', np.NaN,inplace=True)
df1_archive_enhanced_clean.dog_level.replace('doggopuppo', np.NaN, inplace=True)
df1_archive_enhanced_clean.dog_level.replace('doggofloofer', np.NaN, inplace=True)
```

```python
#import libaries necessaries libaries for visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#import twitter_master data set into juypter notebook
df=pd.read_csv('twitter_archive_enhanced_master')
```

### Insights

- Is there a significant relationship between retweet counts and favorite counts in dog breeds

- Which source has a high number of retweet counts based on the second prediction of dog breeds?

- Which level of dog has the highest favorite counts in third dog prediction?

**1. Is there a significant relationship between retweet counts and favorite counts in dog breeds**

```python
# scatterplot that shows the relationship between retweet and favourite counts of dog breed
plt.figure(figsize=(10,6))
plt.title('Relationship between retweet counts and favorite counts')
sns.scatterplot(x=df['retweet_count'], y=df['favorite_count']);
```

**2. Which source has a high number of retweet counts based on the second prediction of dog breeds?**

```python
#The bar graph of the average retweet accounts in source.
plt.figure(figsize=(10,6))
df.query('p2_dog==True').groupby('source')['retweet_count'].mean().plot(kind='bar')
plt.title('The average number of retweet counts in three sources')
plt.xlabel("Source")
plt.ylabel("Average retweet counts")
plt.legend();
```

**3. Which level of dog has the highest favorite counts in third dog prediction?**

```python
plt.figure(figsize=(10,6))
df.query('p3_dog==True').groupby('dog_level')['favorite_count'].mean().plot(kind='bar')
plt.title('The average number of favorite counts in level of dog')
plt.xlabel("Level of dog")
plt.ylabel("Average favorite counts")
plt.legend();
```
